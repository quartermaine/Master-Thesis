```python
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
path = f'{os.getcwd()}/results_and_plots/15_1_2021/19_derivatives_aug/mc_accuracies_alff_degree_binarize_degree_weighted_dual_regression_eigenvector_binarize_eigenvector_weighted_falff_lfcd_reho_vmhc.npy'
accuracies = np.load(path)
```

```python
plt.plot(accuracies)
plt.ylabel('Accuracies')
plt.title("MC dropout evaluation t=1000 19 derivatives with augmentations")
plt.savefig('MC_dropout_19_derivatives_aug.png')
plt.show()
```

```python
# Density Plot and Histogram of all arrival delays
graph = sns.distplot(accuracies, hist=True, kde=True, #rug = True,
             bins=10, color = 'darkblue',
             hist_kws={'edgecolor':'black', "alpha": 0.2},
             kde_kws={'linewidth': 2,},
            # rug_kws={'color': 'black'}
            )

mean_acc = np.mean(accuracies)
sd_acc = np.std(accuracies)
print(f'Accuracy mean: {mean_acc}|sd: {sd_acc}')
graph.axvline(mean_acc, color='tomato', linestyle='-.', label='mean')
graph.axvline(mean_acc + sd_acc, color='blueviolet', linestyle='-.', label='standard deviation')
graph.axvline(mean_acc - sd_acc, color='blueviolet', linestyle='-.')
graph.legend()
plt.title('Histogram with Density line for 19 derivatives with augmentations')
plt.xlabel('Accuracies')
plt.ylabel('Density')
plt.savefig('histogram_MC_dropout_19_derivatives_aug.png')
```
