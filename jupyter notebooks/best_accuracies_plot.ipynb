```python
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use('seaborn')
```

```python
# intialise data of lists. 
data = {'derivative':[f'19_derivatives\n aug', 
                      '19_derivatives\n no_aug', 
                      '9_derivatives\n aug', 
                      '9_derivatives\n no_aug',
                      'dual_regression\n aug',
                      'dual_regression\n no_aug',
                      'reho aug',
                      'reho no_aug'], 
        'best_accuracy':[0.687577, 0.61614, 0.619506, 0.613904, 0.562173, 0.548692, 0.635519, 0.5855577],
        'loss':[1.346517, 1.025252, 1.287407, 0.887029, 1.56641, 1.636199, 0.801207, 1.041307]} 
  
# Create DataFrame 
df = pd.DataFrame(data) 
  
# Print the output. 
df
```

```python
title_plot = 'Plot of accuracy and loss obtained for best architecture'

fig, ax = plt.subplots(figsize = ( 18 , 14 )) 

ax.yaxis.set_tick_params(labelsize='x-large', rotation=0)

ax.xaxis.set_tick_params(labelsize='x-large')

# Plot the scatterplot 
ax = sns.barplot(x="derivative", 
                 y="best_accuracy", 
                 data=df,
                 color="salmon", 
                 saturation=.5)
  
# Set label for x-axis 
ax.set_xlabel( "Accuracy" , size = 22 ) 
  
# Set label for y-axis 
ax.set_ylabel( "Derivative combination" , size = 22 ) 
  
# Set title for plot 
ax.set_title( title_plot , size = 28 ) 

name = 'feature_importances_dual_regression_derivative_no_aug'
# plt.savefig(name)

# Display figure 
plt.show() 
```

```python
df_melt = pd.melt(df, id_vars="derivative", var_name="score", value_name="value")
df_melt
```

```python
title_plot = 'Plot of accuracy and loss obtained for best architecture'

fig, ax = plt.subplots(figsize = ( 20 , 16)) 

ax.yaxis.set_tick_params(labelsize='x-large', rotation=0)

ax.xaxis.set_tick_params(labelsize='x-large')

# Plot the scatterplot 
ax = sns.barplot(x='derivative', y='value', hue='score', data=df_melt)

# Set label for x-axis 
ax.set_ylabel( "Accuracy/Loss" , size = 22 ) 
  
# Set label for y-axis 
ax.set_xlabel( "Derivative combination" , size = 22 ) 
  
# Set title for plot 
ax.set_title( title_plot , size = 28 ) 

plt.setp(ax.get_legend().get_texts(), fontsize='18') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='22') # for legend title

name = 'best_accuracy_plot'
plt.savefig(name)

# Display figure 
plt.show() 

```
